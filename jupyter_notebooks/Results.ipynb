```python
from urllib.parse import urlencode, quote_plus
import requests
import pandas as pd
import numpy as np
from io import StringIO

names = [
    'Исаков Илья Владимирович',
    'Мелик-Заде Сергей Алексеевич',
    'Барановский Александр Александрович',
    'Иван',
    'Сотников Игорь Олегович',
    'Степанов Василий Владимирович',
    'Хасанов Тимур Равильевич',
    'Предеин Сергей',
]

base_url="https://docs.google.com/spreadsheets/d"
sheet_key="1NSdcHzMy_KoFx6hSZUuXUTwW5X6xsFY-POW2wQ4jjjs"
sheet_id="1465042576"
service="gviz/tq?"
query="select * where (" + " OR ".join([("A='" + name + "'") for name in names]) +")"

payload = {'tqx':'out:csv',
           'sheet':sheet_id,
           'tq':query}
encoded = urlencode(payload, quote_via=quote_plus)
sheet_url = "/".join([base_url,sheet_key,service]) + encoded

#print (sheet_url)
r = requests.get(url=sheet_url)
#print(r.text)

df = pd.read_csv(StringIO(r.text))
df = df[np.hstack((['ФИО'], df.columns.values[1:][df.describe().loc['count'] > 0]))]
df.set_index('ФИО')
```
